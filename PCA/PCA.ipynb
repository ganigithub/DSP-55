### PCA : Principal Component Analysis

```python
from sklearn import datasets
import pandas as pd
```

```python
#1.Import iris dataset
iris = datasets.load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['Target'] = iris.get('target')
df.head()
```

```python
#2. create features and target datasets

features = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
x = df[features].values
y = df.Target.values
```

```python
from sklearn.preprocessing import StandardScaler

#3. standardize the features
x = StandardScaler().fit_transform(x)

#preview x
pd.DataFrame(data = x, columns = features).head()
```

```python
#4. Import PCA from sklearn
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)   #4 because of 4 features. We can reduce n_components if we want

#fit PCA to features
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
#5. calculate the variance explained by principle components

print('Variance of each component:', pca.explained_variance_ratio_)  #this shows how much a feature explains variance in data 
print('\n Total Variance Explained:', round(sum(list(pca.explained_variance_ratio_))*100, 2))  #its 100 coz we considered all the 4 featres
```

```python
#6. create a dataset from principal components

df = pd.DataFrame(data = principalComponents, columns = ['PC1','PC2', 'PC3','PC4'])

target    = pd.Series(iris['target'], name = 'target')
result_df = pd.concat([df, target], axis= 1)
result_df
```
