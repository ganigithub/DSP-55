### Data visualization

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
yield_apples = [0.222, 0.57, 0.99, 1.4, 0.99, 2]
```

```python
#line charts

plt.plot(yield_apples)
plt.show()
```

```python
yield_apples = [0.222, 0.57, 0.99, 1.4, 0.99, 2]
years = [1,2,3,4,5,6]

plt.plot(years, yield_apples, color = 'purple')

plt.xlabel('Years', size = 15)
plt.ylabel('Apple Yield', size = 15)
plt.title('Apple Yield 2020-2021', size = 20,  color = 'red')

plt.show()
```

```python
yield_apples = [0.222, 0.57, 0.99, 1.4, 0.99, 1.6]
yield_bananas = [2, 1.9, 1.5, 1, 0.5, 0.4]
years = [1,2,3,4,5,6]

plt.plot(years, yield_apples, color = 'red')
plt.plot(years, yield_bananas, color = 'Yellow')

plt.xlabel('Years', size = 15)
plt.ylabel('Yield', size = 15)
plt.title('Yield 2020-2021', size = 20)

plt.legend(['Apples', 'Bananas'])

plt.show()
```

```python
yield_apples = [0.222, 0.57, 0.99, 1.4, 0.99, 1.6]
yield_bananas = [2, 1.9, 1.5, 1, 0.5, 0.4]
years = [1,2,3,4,5,6]

plt.figure(figsize = (10,10))  #size of the whole graph

plt.plot(years, yield_apples, color = 'red', marker = '*', ls = ':', lw = 2)   #ls = line style, lw = line width
plt.plot(years, yield_bananas, color = 'blue', marker = 'o', ls = '--', lw = 4)

plt.title('yield 20-21')
plt.xlabel('Years')
plt.ylabel('Yield')
plt.legend(['apples','bananas'])

plt.show()
```

### scatter plot using matplotlib

```python
flowers_df = sns.load_dataset('iris')
flowers_df.head()
```

```python
flowers_df.species.unique()
```

```python
flowers_df['sepal_length']
```

```python
flowers_df['sepal_width']
```

```python
#using matplotlib
plt.figure(figsize = (8,8))

plt.scatter(flowers_df['sepal_length'], flowers_df['sepal_width'])

plt.title('length vs width')
plt.xlabel('sepal length')
plt.ylabel('sepal width')

plt.show()
```

### scatterplot using seaborn

```python
plt.figure(figsize = (8,8))

sns.scatterplot(x = flowers_df['sepal_length'],
               y = flowers_df['sepal_width'],
               s = 50)
plt.show()

#no need to put xlables and ylabels in seaborn
```

```python
#introducing third parameter

plt.figure(figsize = (8,8))
sns.scatterplot(x = flowers_df['sepal_length'],
                y = flowers_df['sepal_width'],
                hue = flowers_df['species'],
                s = 100)

plt.show()
```

## Histograms

```python
sns.set_style('darkgrid')  #background
plt.figure(figsize=(8,8))

plt.hist(flowers_df['sepal_width'])
plt.show()
```

```python
#bins

sns.set_style('whitegrid')  #background
plt.figure(figsize=(8,8))

plt.hist(flowers_df['sepal_width'], bins = 6)  #bins will decide no of buckets
plt.show()
```

```python
#comparing two distributions of setose, versicolor

setosa_width = flowers_df[flowers_df['species'] == 'setosa']['sepal_width']

versicolor_width = flowers_df[flowers_df['species'] == 'versicolor']['sepal_width']

sns.set_style('darkgrid')
plt.figure(figsize = (8,8))

plt.hist(setosa_width , bins = 6, alpha = 0.5)  #alpha is for opacity
plt.hist(versicolor_width, bins = 6, alpha = 0.3)
plt.legend(['setosa', 'versicolor'])

plt.show()
```

### Barplot

```python
years = range(2015, 2021)
yield_apples = [0.222, 0.57, 0.99, 1.4, 0.99, 1.6]
yield_bananas = [2, 1.9, 1.5, 1, 0.5, 2]
```

```python
plt.bar(years, yield_bananas, color = 'yellow')

plt.xticks(years, ['year 1', 'year 2', 'year 3', 'year 4', 'year 5', 'year 6'])

plt.figure(figsize = (8,8))
plt.show()
```

```python
y_pos1 = [1,4,7,10,13] #setting gaps between two bars
y_pos2 = [2,5,8,11,14]
y_pos3 = [1.5, 4.5, 7.5, 10.5, 13.5]  #for labeling

bars = ('a', 'b','c','d','e')
height1 = [3,12,5,8,47]
height2 = [4,7,12,10,35]

plt.bar(y_pos1, height1)
plt.bar(y_pos2, height2)

plt.xticks(y_pos3, bars)

plt.show()
```

```python
#remove the column that only contains missing values
df = df.drop(columns = 'county_name')
```

```python
df.head()
```

```python
#who overspeeds men or women?

df[df['driver_gender']=='M']['violation'].value_counts(normalize = True)

#52% of men are stopped for overspeeding
```

```python
df[df['driver_gender']=='F']['violation'].value_counts(normalize = True)

#65% of women are stopped for overspeeding
```

```python
#or using groupby

df.groupby('driver_gender')['violation'].value_counts(normalize = True)

#women overspeed more than men
```

```python
#Does gender affect who gets searched during a stop
df.groupby('driver_gender')['search_conducted'].value_counts(normalize = True)
```

```python
# or
df.groupby('driver_gender')['search_conducted'].mean()
```
